```python
# even_sum = 0
# for number in range(2, 101, 2):
#   print(number)
#   even_sum += number
# print(even_sum)

even_sum = 0
for number in range(2,101,2):
    # print(number)
    even_sum += number   # even_sum = 0 + number 

print(even_sum)    
```

```python
# 2nd way
alternative_sum = 0
for number in range(1, 101):
  if number % 2 == 0:
    # print(number)
    alternative_sum += number
print(alternative_sum)
```

```python
# 3rd way --odd numbers


alternative_sum = 0

for number in range(0,100):
    if number % 2 != 0:
        # print(number)
        alternative_sum += number
print(alternative_sum)
```
